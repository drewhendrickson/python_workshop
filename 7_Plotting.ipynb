# Lecture 7: Plotting using matlibplot

```python
#%pylab inline
#%matplotlib [inline|qt|osx|gtx]
%matplotlib inline

import numpy
import matplotlib.pyplot as plt
```

```python
# building a simple plot

x = numpy.linspace(0, 20, 1000)  # 100 evenly-spaced values from 0 to 50
y = numpy.sin(x)

plt.plot(x, y)
#plt.show()
```

```python
# change the x and y limits

plt.plot(x, y)

plt.xlim(5, 15)
plt.ylim(-1.2, 1.2)
```

```python
# add title and axis labels

plt.plot(x, y)

plt.xlabel('X axis!')
plt.ylabel('Y axis!')
plt.title('Oscillatory!')
```

```python
# change the text properties

plt.plot(x, y)

plt.xlabel('X axis!', fontsize=24, color='green')
plt.ylabel('Y axis!', fontsize=14, color='red')
plt.title('Oscillatory!', fontsize=32, color='pink')
```

```python
# add laTex-based labels

plt.plot(x, y)

plt.title(r'$\sin(x)$', fontsize=38)  # the `r` before the string indicates a "raw string"
```

```python
# change the line color and type

plt.plot(x, y, '--r')
```

```python
# plot two lines with different calls

x = numpy.linspace(0, 20, 1000)
y1 = numpy.sin(x)
y2 = numpy.cos(x)

plt.plot(x, y1, '-b', label='sine')
plt.plot(x, y2, '-r', label='cosine')
plt.ylim(-1.5, 2.0)
```

```python
# add a legend

x = numpy.linspace(0, 20, 1000)
y1 = numpy.sin(x)
y2 = numpy.cos(x)

plt.plot(x, y1, '-b', label='sine')
plt.plot(x, y2, '-r', label='cosine')
plt.ylim(-1.5, 2.0)
plt.legend(loc='upper right')
```

```python
# specify label text

x1 = numpy.linspace(0, 10, 20)
y1 = numpy.sin(x1)

x2 = numpy.linspace(0, 10, 1000)
y2 = numpy.sin(x2)

plt.plot(x1, y1, 'bo', label='sampled')
plt.plot(x2, y2, ':k', label='continuous')
plt.legend()

plt.ylim(-1.5, 2.0)
```

```python
# plotting lines are just linear interpolation between points
# so to plot a smooth line (line above) generate lots of points
# less points result in a choppy line

x1 = numpy.linspace(0, 10, 4)
y1 = numpy.sin(x1)

x2 = numpy.linspace(0, 10, 1000)
y2 = numpy.sin(x2)

plt.plot(x1, y1, 'bo', label='sampled')
plt.plot(x2, y2, ':k', label='continuous')
plt.legend()

plt.ylim(-1.5, 2.0)
```

## Sub-plots

```python
# creating sub-plots

def f(t):
    return numpy.exp(-t) * numpy.cos(2 * numpy.pi * t)

t1 = numpy.arange(0.0, 5.0, 0.1)

plt.figure(1)
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t1, f(t1), '-k')

plt.subplot(212)
plt.plot(t1, numpy.cos(2 * numpy.pi * t1), 'r--')
```

The `figure()` command above is optional because `figure(1)` will be created by default, just as a `subplot(111)` will be created by default if you don’t manually specify any axes. 

The `subplot()` command specifies the number of rows, the number of columns, and the current figure number. Figure number ranges from 1 to (#rows x #cols). 

The commas in the subplot command are optional if the number of figures is less than 10. So `subplot(211)` is identical to `subplot(2, 1, 1)`. 
T
You can create an arbitrary number of subplots and axes. If you want to place an axes manually, i.e., not on a rectangular grid, use the axes() command, which allows you to specify the location as axes([left, bottom, width, height]) where all values are in fractional (0 to 1) coordinates. 

You can create multiple figures by using multiple figure() calls with an increasing figure number. Of course, each figure can contain multiple axes and subplots.

```python
# using figure to make multiple sub-plots

x1 = numpy.arange(0.0, 5.0, 0.05)
x2 = numpy.arange(0.0, 100, 1)


plt.figure(1)                # the first figure
plt.subplot(211)             # the first subplot in the first figure
plt.plot(x1, f(x1))
plt.subplot(212)             # the second subplot in the first figure
plt.plot(x1, 1-2*f(x1))


plt.figure(2)                # a second figure
plt.plot(x2, numpy.sin(x2))    # creates a subplot(111) by default
plt.xlim(0,100)

plt.figure(1)                # figure 1 current; subplot(212) still current
plt.subplot(211)             # make subplot(211) in figure1 current
plt.title('So many sub-plots,\nPlotting is just like Game of Thrones!') 
```

## Other types of plots

```python
# scatter plot

samples = 20
x = numpy.arange(samples)
y = 10 + 5*x + 10*numpy.random.normal(size=samples)

plt.plot(x, y, 'bo')
```

```python
# add a best fitting linear regression scatter plot

# fit linear regression using np.polyfit
m, b = numpy.polyfit(x, y, 1)

plt.plot(x, y, 'bo')
plt.plot(x, m*x + b, '-r')
```

```python
# making a histogram

mu, sigma = 100, 15
x = mu + sigma * numpy.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75)

plt.xlabel('Simulated IQ')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
```

```python
# bar charts

n_groups = 5

means_men = (20, 35, 30, 35, 27)
means_women = (25, 32, 34, 20, 25)

index = numpy.arange(n_groups)
bar_width = 0.35

rects1 = plt.bar(index, means_men, bar_width,
                 color='b',
                 label='Men')

rects2 = plt.bar(index + bar_width, means_women, bar_width,
                 color='r',
                 label='Women')

plt.xlabel('Group')
plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(index + bar_width / 2, ('A', 'B', 'C', 'D', 'E'))
plt.legend()
```

```python
# add opacity and error bars

n_groups = 5

means_men = (20, 35, 30, 35, 27)
std_men = (2, 3, 4, 1, 2)

means_women = (25, 32, 34, 20, 25)
std_women = (3, 5, 2, 3, 3)

index = numpy.arange(n_groups)
bar_width = 0.35

opacity = 0.4
error_config = {'ecolor': '0.3'} # genrating a warning

rects1 = plt.bar(index, means_men, bar_width,
                 alpha=opacity,
                 color='b',
                 yerr=std_men,
                 error_kw=error_config,
                 label='Men')

rects2 = plt.bar(index + bar_width, means_women, bar_width,
                 alpha=opacity,
                 color='r',
                 yerr=std_women,
                 error_kw=error_config,
                 label='Women')

plt.xlabel('Group')
plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(index + bar_width / 2, ('A', 'B', 'C', 'D', 'E'))
plt.legend()
```

```python
# make a series of box plots

data = numpy.random.lognormal(size=(37, 4), mean=1.5, sigma=1.75) + 10
labels = list('ABCD')
fs = 10  # fontsize

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(6, 6), sharey=True)
# shorthand for:
# fig = plt.figure()
# ax = fig.add_subplot(111)


axes[0, 0].boxplot(data, labels=labels)
axes[0, 0].set_title('Default', fontsize=fs)

axes[0, 1].boxplot(data, labels=labels, showmeans=True)
axes[0, 1].set_title('showmeans=True', fontsize=fs)

axes[0, 2].boxplot(data, labels=labels, showmeans=True, meanline=True)
axes[0, 2].set_title('showmeans=True,\nmeanline=True', fontsize=fs)

axes[1, 0].boxplot(data, labels=labels, showbox=False, showcaps=False)
tufte_title = 'Tufte Style \n(showbox=False,\nshowcaps=False)'
axes[1, 0].set_title(tufte_title, fontsize=fs)

axes[1, 1].boxplot(data, labels=labels, notch=True, bootstrap=10000)
axes[1, 1].set_title('notch=True,\nbootstrap=10000', fontsize=fs)

axes[1, 2].boxplot(data, labels=labels, showfliers=False)
axes[1, 2].set_title('showfliers=False', fontsize=fs)

for ax in axes.flatten():
    ax.set_yscale('log')
    ax.set_yticklabels([])

fig.subplots_adjust(hspace=0.4)
```

The line `fig, axes = plt.subplots()` is generating both a figure object and a matrix of Axis objects. The figure is the overall object and the Axis object contains the specific objects plotted in the sub-plot. If there is only one sub-plot, the Figure object can possess the objects.

```python
# arbitrary paths

import matplotlib.path as mpath
import matplotlib.patches as mpatches

fig, ax = plt.subplots()

Path = mpath.Path
path_data = [
    (Path.MOVETO, (1.58, -2.57)),
    (Path.CURVE4, (0.35, -1.1)),
    (Path.CURVE4, (-1.75, 2.0)),
    (Path.CURVE4, (0.375, 2.0)),
    (Path.LINETO, (0.85, 1.15)),
    (Path.CURVE4, (2.2, 3.2)),
    (Path.CURVE4, (3, 0.05)),
    (Path.CURVE4, (2.0, -0.5)),
    (Path.CLOSEPOLY, (1.58, -2.57)),
    ]
codes, verts = zip(*path_data)
path = mpath.Path(verts, codes)
patch = mpatches.PathPatch(path, facecolor='r', alpha=0.5)
ax.add_patch(patch)

# plot control points and connecting lines
x, y = zip(*path.vertices)
line, = ax.plot(x, y, 'go-')

ax.grid()
ax.axis('equal')
```

```python
# plots with different scales

plt.subplots_adjust(hspace=0.4)
t = numpy.arange(0.01, 20.0, 0.01)

#####
# log y axis
plt.subplot(221)
plt.semilogy(t, numpy.exp(-t/5.0))
plt.title('semilogy')
plt.grid(True)

#####
# log x axis
plt.subplot(222)
plt.semilogx(t, numpy.sin(2 * numpy.pi * t))
plt.title('semilogx')
plt.grid(True)

#####
# log x and y axis
plt.subplot(223)
plt.loglog(t, 20 * numpy.exp(-t/10.0), basex=2)
plt.grid(True)
plt.title('loglog base 2 on x')

#####
# with errorbars: clip non-positive values
ax = plt.subplot(224)
ax.set_xscale("log", nonposx='clip')
ax.set_yscale("log", nonposy='clip')

x = 10.0**numpy.linspace(0.0, 2.0, 20)
y = x**2.0

plt.errorbar(x, y, xerr=0.1*x, yerr=5.0 + 0.75*y)
ax.set_ylim(ymin=0.1)
ax.set_title('Errorbars go negative')
```
